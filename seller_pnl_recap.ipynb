# SELLER P&L ANALYSIS

üéØ The goal of this exercise is to compute the P&L associated to each seller  

For each `seller_id` we need three items:  

- The **revenue**:
 - 10% fee on sales
 - 80 BRL per month on Olist


- The **cost**:
 - Review cost according to `{1: 100, 2: 50, 3: 40, 4: 0, 5: 0}` with review score as key and cost in BRL as value


- The **profit** made by Olist

üí° Let's not start from scratch  
‚ùì Import your seller training set and investigate what you already have.


```python
from olist.seller import Seller
sellers = Seller().get_training_data()
```

```python
new_sellers = sellers.drop(['revenues','profit'], axis=1).copy()
```

```python
new_sellers
```

‚ùì What is missing?  
Write down a strategy to get your missing columns  
Re-use as much of what has already been coded in `seller.py` as possible

```python
new_sellers['revenue'] = 0.1 * new_sellers['sales'] +  new_sellers['months_on_olist'] * 80
```

```python
new_sellers.head()
```

```python
orders = Seller().order.get_review_score()
```

```python
from olist.order import Order
orders = Order()
```

```python
orders['cost'] = orders['review_score'].map({1: 100, 2: 50, 3: 40, 4: 0, 5: 0})
```

```python
orders.head()
```

```python
from olist.data import Olist
order_items = Olist().get_data()['order_items']
```

```python
costs = new_sellers.merge(order_items, on="seller_id")\
.merge(orders, on="order_id")[['order_id', 'seller_id', 'cost']]\
.drop_duplicates()\
.groupby('seller_id').sum().copy()
```

```python
new_sellers = new_sellers.merge(costs, on="seller_id")
```

```python
new_sellers['profit2'] = new_sellers['revenue'] -  new_sellers['cost']
```

```python
new_sellers.head()
```

```python
sorted_sellers = new_sellers.sort_values(by="profit2").reset_index()
```

```python
sorted_sellers['profit2'].plot()
```
